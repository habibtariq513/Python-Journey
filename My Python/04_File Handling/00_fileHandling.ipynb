# **usage with +**

   -  ***r+***  | Read   | Overwrite   | No Truncate  | Ptr Start
   -  ***w+***  | Read   | Overwrite   | Truncate     | Ptr End  
   -  ***a+***  | Read   | Append      | No Truncate  | Ptr End

# r+

```python
x = open("02_file_reading.txt", "a")
x.write(" And i'm using for practice purpose so that i can learn it better")
x.close() 

#  yahan pr jo appended statemnt hai wo overwrite ho jaye gi us txt se jo ap ne "r+" k mode k sath likhi hui ho ga
f = open("02_file_reading.txt", "r+")
f.write("This is new file for r+ mode")
print(f.read())
f.close()
                                                 
f = open("02_file_reading.txt", "r")
print("\n",f.read())
```

# w+

```python
# ================= For w+ mode ===================
# agr file exist krti hai to uske content ko erase kr deta h or agr file nahi hai to nayi file create kr deta h 
f = open("w+_file.txt", "w+") 
f.write("Hi I'm pursuing Data Science")
f.seek(0) # move the cursor to the beginning of the file
print(f.read()) 

f = open("w+_file.txt", "a+") 
f.write("\nAnd I'm also learning AI and ML")
f.seek(0) 
print(f.read())

# f = open("w+_file.txt", "r")
f.close()
```

# a+

```python
f = open("append_file.txt", "a+")
f.write("\nI like to learn File Handling")
f.seek(0)
print(f.read()) # ye print nahi kare ga kuch kyun k cursor end pr hai
f.close()
```

# With keyword

```python
with open("file_writing.txt", "w+") as f:
    f.write("Now i'm using WITH keyword")
    f.seek(0)
    print(f.read())
```

# Delete keyword


```python
import os
with open("de.txt", "w") as f:
    f.write("This file is to be deleted")
os.remove("de.txt")
```
